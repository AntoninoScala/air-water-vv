```python
import numpy as np
#np.loadtxt?
tp_exp = np.loadtxt("dambreak_exp_pr1.txt")
tp_pro = np.loadtxt("pressure.txt")
exp_ful = np.loadtxt("exp_full.dat")
exp_height = np.loadtxt("exp_height.dat")
exp_pres = np.loadtxt("exp_pres.dat")
```

```python
%matplotlib notebook
#import matplotlib
from matplotlib import pyplot as plt
fig, axes = plt.subplots(2,2)
axes[0,0].plot(tp_pro[:,0], tp_pro[:,1])
axes[0,0].plot(exp_pres[:1500,0], exp_pres[:1500,1],'--')
axes[0,1].plot(tp_pro[:,0], tp_pro[:,2])
axes[0,1].plot(exp_pres[:1500,0], exp_pres[:1500,2],'--')
axes[1,0].plot(tp_pro[:,0], tp_pro[:,3])
axes[1,0].plot(exp_pres[:1500,0], exp_pres[:1500,3],'--')
axes[1,1].plot(tp_pro[:,0], tp_pro[:,4])
axes[1,1].plot(exp_pres[:1500,0], exp_pres[:1500,4],'--')
from matplotlib.ticker import MultipleLocator, FormatStrFormatter
minorLocator = MultipleLocator(.05)
for i in range(2):
    for j in range(2):
        axes[i,j].xaxis.set_minor_locator(minorLocator)
        axes[i,j].xaxis.grid(True, which='minor')
fig.savefig("pressure_comp.svg")
```
