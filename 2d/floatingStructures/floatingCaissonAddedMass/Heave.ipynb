```python
%matplotlib notebook
import numpy as np
from matplotlib import pyplot as plt
```

```python
caisson_am=np.loadtxt("record_caisson_am.csv",delimiter=",",skiprows=1)
caisson=np.loadtxt("record_caisson.csv",delimiter=",",skiprows=1)
caisson_am_nc=np.loadtxt("record_caisson_am_nc.csv",delimiter=",",skiprows=1)
caisson_nc=np.loadtxt("record_caisson_nc.csv",delimiter=",",skiprows=1)
caisson_chrono=np.loadtxt("record_caisson_chrono.csv",delimiter=",",skiprows=1)
caisson_chrono_am=np.loadtxt("record_caisson_chrono_am.csv",delimiter=",",skiprows=1)
caisson_chrono_am_nc=np.loadtxt("record_caisson_chrono_am_nc.csv",delimiter=",",skiprows=1)
caisson_chrono_nc=np.loadtxt("record_caisson_chrono_nc.csv",delimiter=",",skiprows=1)
```

```python
fig = plt.figure()
plt.plot(caisson[:,0],caisson[:,2],)
plt.plot(caisson_am[:,0],caisson_am[:,2])
plt.plot(caisson_nc[:,0],caisson_nc[:,2])
plt.plot(caisson_am_nc[:,0],caisson_am_nc[:,2])
plt.plot(caisson_chrono[:,0],caisson_chrono[:,4])
plt.plot(caisson_chrono_am[:,0],caisson_chrono_am[:,4])
plt.plot(caisson_chrono_am_nc[:,0],caisson_chrono_am_nc[:,4])
plt.plot(caisson_chrono_nc[:,0],caisson_chrono_nc[:,4])
plt.legend(['c','c+am','nc','nc+am','CH+c','CH+c+am','CH+nc','CH+nc+am'])
```
