```python
%matplotlib notebook
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
ball_position = pd.read_csv("./res-1.csv", sep='\s+',header=None)
```

```python
print(ball_position)
```

```python
ball_position = ball_position.values
```

```python
ball0 = ball_position[0::3,:]
ball1 = ball_position[1::3,:]
ball2 = ball_position[2::3,:]
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
for i in range(ball0.shape[0]):
    ax.add_patch(plt.Circle((ball0[i,0],ball0[i,1]), 0.1, color='r', alpha=0.1))
    ax.add_patch(plt.Circle((ball1[i,0],ball1[i,1]), 0.1, color='b', alpha=0.1))
    ax.add_patch(plt.Circle((ball2[i,0],ball2[i,1]), 0.1, color='g', alpha=0.1))

ax.set_aspect('equal', adjustable='datalim')
ax.plot()   #Causes an autoscale update.
plt.show()
```
